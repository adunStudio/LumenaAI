```python
from infrastructure.database.mongo_client import MongoDBClient
from infrastructure.repository import YoutubeContentRepository
from infrastructure.repository import YoutubeScriptCollectionRepository

from domain import YouTubeContent, YoutubeScriptCollection

import os
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.environ['OPENAI_API_KEY']

# MongoDB 클라이언트 초기화
client = MongoDBClient(uri=os.environ['MONGO_CONNECTION_STRING'])
client.connect()

# 저장소 초기화
youtube_content_repository = YoutubeContentRepository(client)
youtube_script_collection_repository = YoutubeScriptCollectionRepository(client)



contents = youtube_content_repository.find_all()

count = len(contents)

```

```python
for i, content in enumerate(contents):
    print(1)
    #content: YouTubeContent
    if youtube_script_collection_repository.get(content.url.url) is None:
        print(f'없다: {i + 1}/{count}: {content.channel}: {content.title}: {content.url.url}')

        youtube_script_collection = YoutubeScriptCollection(content.url.url)
        youtube_script_collection.set_auto_script(content.script_auto)
        youtube_script_collection.set_whisper_script(content.script_whisper)
        youtube_script_collection.set_refined_script(content.script)

        youtube_script_collection_repository.save(youtube_script_collection)
    else:
        print(f'있다: {i + 1}/{count}: {content.channel}: {content.title}')
```

```python
for i, content in enumerate(contents):
    print(f'{i + 1}/{count}: {content.channel}: {content.title}: {content.url.url}')
    #content: YouTubeContent
    content.set_script(None)
    content.set_script_auto(None)
    content.set_script_whisper(None)
    youtube_content_repository.save(content)
```
